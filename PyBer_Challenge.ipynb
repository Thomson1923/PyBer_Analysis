```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)

#ride_data_df["date"]=pd.to_numeric(ride_data_df["date"])

ride_data_df.dtypes
```

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

```python
#  1. Get the total rides for each city type
pyber_data_df.set_index(["type"])
total_ride_count = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_ride_count
```

```python
# 2. Get the total drivers for each city
total_driver_count = city_data_df.groupby(["type"]).sum()["driver_count"]
total_driver_count
```

```python
#  3. Get the total amount of fares for each city type
total_fare_sum = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fare_sum
```

```python
#  4. Get the average fare per ride for each city type. 
average_fare_per_ride=pyber_data_df.groupby(["type"]).mean()["fare"]
average_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type.
average_fare_per_driver=pyber_data_df.groupby(["type"]).sum()["fare"] / total_driver_count
average_fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. 

pyber_summary_df = pd.merge(total_ride_count,total_driver_count, how="left", on=["type"])
pyber_summary_df = pd.merge(pyber_summary_df, total_fare_sum, how="left",on=["type"])

pyber_summary_df=pyber_summary_df.rename(columns={"ride_id":"Total Rides","driver_count":"Total Drivers","fare":"Total Fares"})

pyber_summary_df["Average Fare per Ride"]=pyber_data_df.groupby(["type"]).sum()["fare"] / total_ride_count
# Note: pyber_summary_df = pd.merge(pyber_summary_df, average_fare_per_ride, how="left",on=["type"]) would also work
pyber_summary_df["Average Fare per Driver"]=pyber_data_df.groupby(["type"]).sum()["fare"] / total_driver_count

# Display the data frame
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
```

```python
#  8. Format the columns.
pyber_summary_df["Total Rides"]=pyber_summary_df["Total Rides"].map("{:,}".format)
pyber_summary_df["Total Drivers"]=pyber_summary_df["Total Drivers"].map("{:,}".format)
pyber_summary_df["Total Fares"]=pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Ride"]=pyber_summary_df["Average Fare per Ride"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Driver"]=pyber_summary_df["Average Fare per Driver"].map("${:,.2f}".format)

pyber_summary_df
```
